## 13.📊 Views vs Likes (Log Scale)


```python
 

plt.figure()
plt.scatter(df['view_count'], df['like_count'])
plt.xscale('log')
plt.yscale('log')
plt.xlabel('View Count (log scale)')
plt.ylabel('Like Count (log scale)')
plt.title('Views vs Likes (Log Scale)')
plt.show()

```

## 14.📊 Top 5 Most Viewed YouTube Videos


```python
top5 = df.sort_values('view_count', ascending=False).head(5)

plt.figure()
plt.bar(top5['title'], top5['view_count'])
plt.xlabel('Video Title')
plt.ylabel('View Count')
plt.title('Top 5 Most Viewed Videos')

for i, v in enumerate(top5['view_count']):
    plt.text(i, v, f'{v:,}', ha='center', va='bottom')

plt.xticks(rotation=30, ha='right')
plt.show()
```

## 15.📈 Monthly Total Views Trend


```python
monthly_views = df.groupby('published_month')['view_count'].sum()

plt.figure()
plt.plot(monthly_views.index, monthly_views.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Total Views')
plt.title('Monthly Total Views Trend')

# Annotate peak month
peak_month = monthly_views.idxmax()
peak_value = monthly_views.max()

plt.annotate(
    f'Peak: {peak_value:,}',
    xy=(peak_month, peak_value),
    xytext=(peak_month, peak_value * 0.7),
    arrowprops=dict(arrowstyle='->')
)

plt.show()
```
